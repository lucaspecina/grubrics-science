## PrimeIntellect - SYNTHETIC-1 - Preference

https://huggingface.co/datasets/PrimeIntellect/SYNTHETIC-1-Preference-Data/viewer

```python
import pandas as pd

# Login using e.g. `huggingface-cli login` to access this dataset
df = pd.read_parquet("hf://datasets/PrimeIntellect/SYNTHETIC-1-Preference-Data/data/train-00000-of-00001.parquet")
```

```python
print(df.info())
# df.head()
```

```python
df.task_type.value_counts()
```

```python
df_llm = df[df.task_type=='llm_judgeable_groundtruth_similarity']
```

```python
import re

def extract_dotcom_or_net(s):
    # Use regex to extract the substring ending with .com or .net (non-greedy)
    if isinstance(s, str):
        match = re.search(r'(https?://[^\s]*?\.(com|net))\b', s)
        if match:
            return match.group(1)
        # fallback: match anything ending in .com or .net, even without http(s)://
        match = re.search(r'([^\s]*?\.(com|net))\b', s)
        if match:
            return match.group(1)
    return None

df_llm['source'] = df_llm['prompt'].apply(extract_dotcom_or_net)
df_llm.source.value_counts(dropna=False)
```

```python
pd.set_option('display.max_colwidth', 200)

print(df_llm[df_llm['source'].isna()][['prompt']])
```
